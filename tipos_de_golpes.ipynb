```python
import pandas as pd

dados = pd.read_csv("respostas-forms-bruto.csv")
dados
```

```python
dados['Qual foi o último tipo de golpe que você presenciou ou foi vítima?'].unique()
```

```python
def freq_relativa_do_golpe(celula):
  total_respostas = len(dados['Qual foi o último tipo de golpe que você presenciou ou foi vítima?'])
  pix_falso = dados['Qual foi o último tipo de golpe que você presenciou ou foi vítima?'].value_counts()['Pix falso']
  golpe_online = dados['Qual foi o último tipo de golpe que você presenciou ou foi vítima?'].value_counts()['Golpe online']
  prod_inex = dados['Qual foi o último tipo de golpe que você presenciou ou foi vítima?'].value_counts()['Compra de produtos inexistente']
  outro = dados['Qual foi o último tipo de golpe que você presenciou ou foi vítima?'].value_counts()['Outro']
  golpe_inexistente = dados['Qual foi o último tipo de golpe que você presenciou ou foi vítima?'].value_counts()['Nunca presenciei e nem fui vítima de golpe']
  fraude_banc = dados['Qual foi o último tipo de golpe que você presenciou ou foi vítima?'].value_counts()['Fraude bancária']

  if celula == 'Pix falso':
    resultado = f"{(pix_falso / total_respostas) * 100:.2f}"
    return float(resultado)
  elif celula == 'Golpe online':
    resultado = f"{(golpe_online / total_respostas) * 100:.2f}"
    return float(resultado)
  elif celula == 'Compra de produtos inexistente':
    resultado = f"{( prod_inex / total_respostas) * 100:.2f}"
    return float(resultado)
  elif celula == 'Outro':
    resultado = f"{(outro / total_respostas) * 100:.2f}"
    return float(resultado)
  elif celula == 'Nunca presenciei e nem fui vítima de golpe':
    resultado = f"{(golpe_inexistente / total_respostas) * 100:.2f}"
    return float(resultado)
  elif celula == 'Fraude bancária':
    resultado = f"{(fraude_banc / total_respostas) * 100:.2f}"
    return float(resultado)


dados["Chance de sofrer golpe em %"] = dados["Qual foi o último tipo de golpe que você presenciou ou foi vítima?"].apply(freq_relativa_do_golpe)
```
